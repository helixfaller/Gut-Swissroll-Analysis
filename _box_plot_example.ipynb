```python
import swissroll as sr
```

```python
import pandas as pd
```

```python
data = pd.read_csv("template.csv", sep=",")
```

```python
## hue is Sacrifice Age (month) and splits Genotype
## hue_split is a variable you use to choose between different
## hue values, but let me describe

hues = []
for variable in data["Sacrifice Age (month)"].tolist():
    if variable not in hues:
        hues.append(variable)

print(f"hue = {hues}")
## if hue_split = 0, then hue == 9
print(f"\nhue[0] = {hues[0]}")
## if hue_split = 1, then hue == 11
print(f"\nhue[1] = {hues[1]}")
## with variable hue and hue_split you can choose which group 
## to test against which
```

```python
sr.box_plot(columns="SIT_Num_TOT", data=data, genotype=["X", "Y"], hue="Sacrifice Age (month)",
            hue_split=None, test="Mann-Whitney", palette="Set2_r")
```
